```python
# 1. Make the following changed to column names:
# lawenforcementagency -> agency
# raceethnicity        -> race

killings.columns = [u'Unnamed: 0', u'name', u'age', u'gender', u'race', u'month',
       u'day', u'year', u'streetaddress', u'city', u'state',
       u'agency', u'cause', u'armed', u'county_income']
killings.head()
```

```python
# 2. Show the count of missing values in each column

killings.isnull().sum()
```

```python
# 3. replace each null value in the dataframe with the string "Unknown"

killcheck = killings['streetaddress'].isnull()

killings['streetaddress'].fillna("Unknown", inplace=True)

killings['streetaddress'][killcheck]
```

```python
# 4. How many killings were there so far in 2015?

killings['year'].value_counts()
```

```python
# 5. Of all killings, how many were male and how many female?

killings.gender.value_counts()
```

```python
# 6. How many killings were of unarmed people?

killings.armed.value_counts()
unarmed = killings.armed=="No"
unarmed.sum()
```

```python
# 7. What percentage of all killings were unarmed?

total_unarmed = unarmed.sum()
total_killed = len(killings)

percent_killed = 100 * total_unarmed / total_killed
percent_killed
```

```python
# 4. Plot the data from the last question in a bar chart, include proper title, and labels!

maj = majors[['Major', 'Median']].sort_values(by='Median').tail(10)

x = range(1,11)
y = maj['Median']
LABELS = maj['Major']

plt.barh(x, y)
plt.yticks(x, LABELS)
plt.title('Majors with the Highest Median Income')
plt.show()
```
