```python
import pandas as pd
killings = pd.read_csv('https://raw.githubusercontent.com/misrab/homework/master/BalaCattavarayane/data/police-killings.csv')
killings.head()
```

```python
# 1. Make the following changed to column names:
# lawenforcementagency -> agency
# raceethnicity        -> race
killings.rename(columns={'lawenforcementagency':'agency', 'raceethnicity':'race'}, inplace=True)
killings.head()
```

```python
# 10. Display a histogram of ages of all killings
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

plt.hist(killings.age,color='red',normed=True)
plt.title("Histogram of Killings by Age")
plt.xlabel("Age")
plt.ylabel("Frequency Distribution")
plt.show()
```

```python
# 11. Show 6 histograms of ages by race
killings.age.hist(by=killings.race, sharex = False, sharey = False ,normed=True)
```

```python
# 12. What is the average age of death by race?
killings.groupby('race').age.mean()
```

```python
# 13. Show a bar chart with counts of deaths every month
monthlyCount = np.array(killings.month.value_counts())
month = np.arange(len(killings.month.unique()))

plt.bar(month,monthlyCount,align='center')
plt.xlabel("Month")
plt.ylabel("Number of Killings")
plt.title("Bar Plot of Number of Killings per Month")
```
