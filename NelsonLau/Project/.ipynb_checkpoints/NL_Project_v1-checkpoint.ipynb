## Making RESPONSE columns

```python
def make_return_column(dataframe,mins_ahead):
    column_name = '%s_min_return' % mins_ahead
    if column_name in dataframe.columns:
        print "Error: %s already in dataframe."
    else:
        dataframe[column_name] = dataframe.resample('T',loffset='-%smin' % mins_ahead,fill_method='ffill',limit=mins_ahead - 1).LAST_PRICE.pct_change(periods=mins_ahead)
```

```python
standard_periods_list = [1,5,10,20,40,80]
for mins_ahead in standard_periods_list:
    make_return_column(one_minute_dataframe, mins_ahead)
return_columns = ['%s_min_return' % x for x in standard_periods_list] 
one_minute_dataframe[return_columns].describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
num_plots = len(standard_periods_list)
f, ax_tuple = plt.subplots(num_plots,sharex='all',sharey='all',figsize=(5,25))
for i in range(num_plots):
    sns.distplot(one_minute_dataframe[one_minute_dataframe['%s_min_return' % standard_periods_list[i]].notnull()]['%s_min_return' % standard_periods_list[i]], ax = ax_tuple[i])
```

```python
return_columns = ['%s_min_return' % x for x in standard_periods_list] 
sns.pairplot(one_minute_dataframe[return_columns].dropna())
```

```python
import scipy.stats as ss
return_columns = ['%s_min_return' % x for x in standard_periods_list] 
result = zip(standard_periods_list, ss.describe(one_minute_dataframe[return_columns].dropna(),axis=0).skewness, ss.describe(one_minute_dataframe[return_columns].dropna(),axis=0).kurtosis)
print "We are using Fisher kurtosis, so a sample from a  normal distribution should report a 0 kurtosis on this measure"
for period, skew, kurt in result:
    print "For %s_min_return, skewness is %f; kurtosis is %f" % (period, skew, kurt)
```

```python
import scipy.stats.mstats
for x in standard_periods_list: 
    k2, pval = scipy.stats.mstats.normaltest(one_minute_dataframe['%s_min_return' % x].dropna())
    print "For %s_min_return, p-value of rejecting hypothesis that sample is from a normal distribution is %f. " % (x, pval)
```

i.e., we observe fat-tailed distributions (as expected).

### Here we make categorical responses. We'll have 1 for a positive return, 0 for a negative return (and NaN for a return that is exactly 0.)

```python
for x in standard_periods_list:
    one_minute_dataframe['%s_min_updown' % x] = one_minute_dataframe['%s_min_return' % x][abs(one_minute_dataframe['%s_min_return' % x]) > 0].apply(lambda x: 1 if x > 0 else 0)
one_minute_dataframe.head()
```

```python
return_columns = ['%s_min_return' % x for x in standard_periods_list]
updown_columns = ['%s_min_updown' % x for x in standard_periods_list]
response_columns = return_columns + updown_columns
one_minute_dataframe[response_columns].describe()
```

```python
for x in standard_periods_list:
    num_positive_continuous = one_minute_dataframe['%s_min_return' % x][one_minute_dataframe['%s_min_return' % x]>0].count()
    num_negative_continuous = one_minute_dataframe['%s_min_return' % x][one_minute_dataframe['%s_min_return' % x]<0].count()
    num_positive_categorical = one_minute_dataframe['%s_min_updown' % x][one_minute_dataframe['%s_min_updown' % x] == 1].count()
    num_negative_categorical = one_minute_dataframe['%s_min_updown' % x][one_minute_dataframe['%s_min_updown' % x] == 0].count()
    print "For %s min returns: +ve: %s cts %s cat ; -ve: %s cts %s cat " % (x, num_positive_continuous, num_positive_categorical, num_negative_continuous, num_negative_categorical)
```
