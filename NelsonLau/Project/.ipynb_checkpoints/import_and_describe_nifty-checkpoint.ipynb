```python
one_minute_dataframe['date_only'] = one_minute_dataframe.Date.dt.to_period('D')
one_minute_dataframe.head()
```

```python
daily_dataframe['date_only'] = daily_dataframe.Date.dt.to_period('D')
daily_dataframe.head()
```

```python
one_minute_dataframe = pd.merge(one_minute_dataframe, daily_dataframe[['FUT_CUR_GEN_TICKER', 'date_only']], on='date_only', how='left')
one_minute_dataframe.head()
```

Now we've included the current futures contract in the one minute dataframe. Bloomberg adjusts the prices for rolls, as we've downloaded NZ1 generic, but we just want to be able to check on roll dates if anything strange happened due to the adjustment

```python
one_minute_dataframe['DayOfWeek'] = one_minute_dataframe.Date.dt.dayofweek
one_minute_dataframe.DayOfWeek.hist(bins=5)
```

```python
%matplotlib inline 
one_minute_dataframe['Hour'] = one_minute_dataframe.Date.dt.hour
one_minute_dataframe['Hour'].hist(bins=8)
```

```python
one_minute_dataframe[one_minute_dataframe['Hour'] == 18].head(10)
```

The NIFTY market is open from 1145-1800. So having fewer rows with hour = 11 makes sense. The small number of rows with hour = 18 seem to be closing and settlement prints.

```python
ts = one_minute_dataframe.set_index('Date')
ts.head()
```

```python
import matplotlib.pyplot as plt
ts['Return_1min'] = (ts.LAST_PRICE - ts.LAST_PRICE.shift(-1, freq = 'min'))/ts.LAST_PRICE.shift(-1, freq = 'min')
plt.hist(ts[(ts['Return_1min'].isnull()) & (ts['Hour'] < 18)].index.time)
```

Most of the NaNs in the 1-min forward are occuring at the end of the day.

```python
for return_period in [1,5,10,20,40,80]:
    ts['Return_%smin' % return_period] = (ts.LAST_PRICE - ts.LAST_PRICE.shift(-return_period, freq = 'min'))/ts.LAST_PRICE.shift(-return_period, freq = 'min')
ts[['Return_1min', 'Return_5min', 'Return_10min', 'Return_20min', 'Return_40min', 'Return_80min']].describe()
```

```python
for return_period in [1,5,10,20,40,80]:
    ts['Prior_%smin' % return_period] = (ts.LAST_PRICE - ts.LAST_PRICE.shift(return_period, freq = 'min'))/ts.LAST_PRICE.shift(return_period, freq = 'min')
ts[['Prior_1min', 'Prior_5min', 'Prior_10min', 'Prior_20min', 'Prior_40min', 'Prior_80min']].describe()
```
