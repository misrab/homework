```python
# 10. Display a histogram of ages of all killings

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

plt.hist(killings.age,color='blue',normed=True)
plt.title("Histogram of Killings by Age")
plt.xlabel("Age")
plt.ylabel("Frequency Density")
plt.show()
```
